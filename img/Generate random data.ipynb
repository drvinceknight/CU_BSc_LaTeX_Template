```python
import matplotlib.pyplot as plt
import seaborn
import random
%matplotlib inline

# Generate some stupid data
a = 2
b = 5

x = range(1, 100)
y = [a * (1 + (random.random() - .5)) * point + b for point in x]
yd = [-a * (1 + (random.random() - .5)) * point + b + 200 for point in x]
```

```python
plt.figure()
plt.scatter(x, y, label='blue dots')
plt.scatter(x, yd, color='red', label='red dots')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Some amazing data')
plt.legend(loc=6)
plt.savefig('plot1.pdf')
plt.show();
```

```python
alpha = .5

plt.figure()
plt.hist(y, label='blue dots', alpha=alpha)
plt.hist(yd, label='red dots', alpha=alpha, color='red')
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Distribution of the amazing data')
plt.legend(loc=2)
plt.savefig('plot2.pdf')
plt.show();
```
